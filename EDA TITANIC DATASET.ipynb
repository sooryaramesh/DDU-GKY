## histogram

```python
sns.histplot(data=data,x='Age')
plt.show()
```

## box plot

```python
sns.boxplot(data=data,x='Age')
plt.show()
```

## KDE plot

```python
sns.kdeplot(data=data,x='Age')
plt.show()
```

## sub plot in matplotlib

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.boxplot(data=data,y='Age')
plt.subplot(2,2,2)
sns.boxplot(data=data,y='Pclass')
plt.subplot(2,2,3)
sns.boxplot(data=data,y='Fare')
plt.subplot(2,2,4)
sns.boxplot(data=data,y='Parch')
plt.show()
```

## line chart

```python
plt.figure(figsize=(10,6))
plt.xlabel('Age')

plt.ylabel('Survived')
sns.lineplot(data=data,x='Age', y='Fare',hue='Sex')
plt.show()
```

## count plot

```python
plt.figure(figsize=(16,5))
plt.xticks(rotation = 90)
sns.countplot(data=data,x='Survived')
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.xticks(rotation = 90)
sns.countplot(data=data,x='Survived',hue='Sex')
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.xticks(rotation = 90)
sns.countplot(data=data,x='Survived',hue='Pclass')
plt.show()
```

## horizontal bar chart

```python
plt.figure(figsize=(6,6))
plt.xticks(rotation=75)
sns.barplot(data=data,x='Survived',y='Age',hue='Sex')
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.xticks(rotation = 90)
sns.countplot(data=data,x='SibSp')
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.xticks(rotation = 90)
sns.countplot(data=data,x='SibSp',hue='Pclass')
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.xticks(rotation = 90)
sns.countplot(data=data,x='Parch')
plt.show()
```

## pie chart

```python
data.Survived.value_counts().index
```

```python
palette_color = sns.color_palette('dark')
```
