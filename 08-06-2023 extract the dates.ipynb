```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('./Car_sales (1).csv',parse_dates=['Latest_Launch'])
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().mean()*100
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data.sort_values(by='Latest_Launch',inplace=True)
```

```python
data.head()
```

```python
data.reset_index(drop=True,inplace=True)
```

## extracting the dates

```python
data["year"]=data.Latest_Launch.dt.year
data["month"]=data.Latest_Launch.dt.month
data["Day"]=data.Latest_Launch.dt.day
```

```python
data.head()
```

## univeriate analysis

```python
sns.histplot(data=data, x='Sales_in_thousands')
plt.show()
```

## box plot

```python
sns.boxplot(data=data, x='Sales_in_thousands')
plt.show()
```

```python
sns.boxplot(data=data, y='Sales_in_thousands')
plt.show()
```

## KDE plot

```python
sns.kdeplot(data=data.Sales_in_thousands)
plt.show()
```

## sub plot

```python
sns.histplot(data=data)
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.xticks(rotation = 75)
sns.boxplot(data=data)
plt.show()
```

```python
sns.kdeplot(data=data)
plt.show()
```

## FOR LOOP TO PLOT SUBPLOT

```python
for i in data.columns:
    if data[i].dtypes!='object':
        sns.boxplot(y=data[i])
        plt.show()
```
