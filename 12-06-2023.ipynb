#### Loading the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### Read the dataset

```python
data=pd.read_csv('./Cuisine_rating.csv')
```

```python
data.info()
```

## Descriptive Statistics

```python
data.describe()
```

```python
data.columns
```

#### Renaming the columns

```python
data.rename(columns={'Food Rating':'FoodRating','Service Rating':'ServiceRating','Overall Rating':'OverallRating','Alcohol ':'Alcohol','Often A S':'OftenAS'},inplace=True)
```

```python
data.isnull().mean()*100
```

#### To create Age column

```python
data['Age']=[(2022-DOB)for DOB in data['YOB']]
```

#### Droping unwanted columns

```python
data.drop(columns=['User ID','Area code','YOB'],inplace=True)
```

### Creating a custom function to filter out location wise datasets

```python
def sort_Location(location):
    data_location = data[data.Location==location]
    data_location.drop(columns='Location',inplace=True)
    return data_location
```

```python
Upper_East_Side_NY=sort_Location('Upper East Side,NY')
```

```python
Upper_East_Side_NY
```

```python
St_George_NY=sort_Location('St. George,NY')
```

```python
St_George_NY
```

## Line chart

```python
plt.figure(figsize=(10,6))
plt.xlabel('Overall Rating')
plt.ylabel('Cuisines')
sns.lineplot(data=Upper_East_Side_NY,x='OverallRating',y='Cuisines')
plt.show()
```

#### In Upper_East_Side_NY city ,Japanese cuisines got the best overall rating and chinese food got the least.

```python
plt.figure(figsize=(10,6))
plt.xlabel('ServiceRating')
plt.ylabel('FoodRating')
sns.lineplot(data=St_George_NY,x='ServiceRating',y='FoodRating')
plt.show()
```

####  In  Upper_East_Side_NY city,there is a positive correlation between servicerating and foodrating
