### EDA


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('./Store Sales.csv',parse_dates=['Ship Date','Order Date'])
```

```python
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().mean()*100
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().mean()*100
```

```python
data.columns
```

```python
data.rename(columns={'Ship Mode':'ShipMode','Customer ID':'CustomerID','Customer Name':'CustomerName','Sub-Category':'SubCategory','Product Name':'ProductName','Ship Date':'ShipDate','Product ID':'ProductID','Order Date':'OrderDate','Postal Code':'PostalCode'},inplace=True)
```

```python
data.columns
```

```python
data["year"]=data.OrderDate.dt.year
data["month"]=data.OrderDate.dt.month
data["Day"]=data.OrderDate.dt.day
```

```python
data["year"]=data.ShipDate.dt.year
data["month"]=data.ShipDate.dt.month
data["Day"]=data.ShipDate.dt.day
```

```python
data['Delivery'] = data['OrderDate']- data['ShipDate']
```

```python
data
```

```python
data.columns
```

```python
data.head(10)
```

```python
data.drop(columns=['Row ID','Order ID','CustomerID','ProductID','PostalCode'],inplace=True)
```

```python
data.head()
```

```python
data['Delivery1']=data.Delivery.astype('str')
```

```python
data
```

```python
data['Delivery1'] = data.Delivery1.str.replace('\D', "", regex=True)
data
```

```python
data['Delivery1'] = data.Delivery1.astype('int') 
```

```python
data.info()
```

```python
data['weekDay']=data.OrderDate.dt.weekday
```

```python
data
```

```python
data.weekDay.unique()
```

```python
data.weekDay.replace({0:'sunday',1:'Monday',2:'Tuesday',3:'wednesday',4:'Thursday',5:'Friday',6:'Saturday'},inplace=True)
```

```python

data
```

```python
def sort_State(state):
    data_state = data[data.State==state]
    data_state.drop(columns='State',inplace=True)
    return data_state
```

```python
California=sort_State('California')
California
```

```python
California.sort_values(by='month', inplace=True)
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('month')
plt.ylabel('Delay')
sns.lineplot(data=California,x='month',y='Delay')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('month')
plt.ylabel('Sales')
sns.lineplot(data=California,x='month',y='Sales',hue='Segment')
plt.xticks(rotation = 90)
plt.show()
```
